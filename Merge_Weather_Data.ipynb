### Merge Tables

```python
#make copies of data for origin and destination cities

origin_jan_2018 = jan2018_weather.copy()
origin_jan_2018 = origin_jan_2018.rename(columns={ 'maxtempC':'OriginmaxtempC', 'mintempC':'OriginmintempC', 
                                                  'totalsnow_cm':'Origintotalsnow_cm', 'sunhour':'Originsunhour',
       'uvIndex':'OriginuvIndex', 'moon_illumination':'Originmoon_illumination', 'moonrise':'Originmoonrise', 
                                                  'moonset':'Originmoonset', 'sunrise':'Originsunrise',
       'sunset':'Originsunset', 'dewpointC':'OrigindewpointC', 'feelslikeC':'OriginfeelslikeC', 'heatindexC':'OriginheatindexC',
                                                  'windchillC':'OriginwindchillC',
       'windgustKmph':'OriginwindgustKmph', 'cloudcover':'Origincloudcover', 'humidity':'Originhumidity', 
                                                  'precipMM':'OriginprecipMM', 'pressure':'Originpressure',
       'tempC':'OrigintempC', 'visibility':'Originvisibility', 'winddirectiondegree':'Originwinddirectiondegree', 
                                                  'windspeedKmph':'OriginwindspeedKmph' })


destin_jan_2018 = jan2018_weather.copy()
destin_jan_2018 = destin_jan_2018.rename(columns={ 'maxtempC':'DestinmaxtempC', 'mintempC':'DestinmintempC', 
                                                  'totalsnow_cm':'Destintotalsnow_cm', 'sunhour':'Destinsunhour',
       'uvIndex':'DestinuvIndex', 'moon_illumination':'Destinmoon_illumination', 'moonrise':'Destinmoonrise', 
                                                  'moonset':'Destinmoonset', 'sunrise':'Destinsunrise',
       'sunset':'Destinsunset', 'dewpointC':'DestindewpointC', 'feelslikeC':'DestinfeelslikeC', 'heatindexC':'DestinheatindexC',
                                                  'windchillC':'DestinwindchillC',
       'windgustKmph':'DestinwindgustKmph', 'cloudcover':'Destincloudcover', 'humidity':'Destinhumidity', 
                                                  'precipMM':'DestinprecipMM', 'pressure':'Destinpressure',
       'tempC':'DestintempC', 'visibility':'Destinvisibility', 'winddirectiondegree':'Destinwinddirectiondegree', 
                                                  'windspeedKmph':'DestinwindspeedKmph' })


origin_jan_2019 = jan2019_weather.copy()
origin_jan_2019 = origin_jan_2019.rename(columns={ 'maxtempC':'OriginmaxtempC', 'mintempC':'OriginmintempC', 
                                                  'totalsnow_cm':'Origintotalsnow_cm', 'sunhour':'Originsunhour',
       'uvIndex':'OriginuvIndex', 'moon_illumination':'Originmoon_illumination', 'moonrise':'Originmoonrise', 
                                                  'moonset':'Originmoonset', 'sunrise':'Originsunrise',
       'sunset':'Originsunset', 'dewpointC':'OrigindewpointC', 'feelslikeC':'OriginfeelslikeC', 'heatindexC':'OriginheatindexC',
                                                  'windchillC':'OriginwindchillC',
       'windgustKmph':'OriginwindgustKmph', 'cloudcover':'Origincloudcover', 'humidity':'Originhumidity', 
                                                  'precipMM':'OriginprecipMM', 'pressure':'Originpressure',
       'tempC':'OrigintempC', 'visibility':'Originvisibility', 'winddirectiondegree':'Originwinddirectiondegree', 
                                                  'windspeedKmph':'OriginwindspeedKmph' })

destin_jan_2019 = jan2019_weather.copy()
destin_jan_2019 = destin_jan_2019.rename(columns={ 'maxtempC':'DestinmaxtempC', 'mintempC':'DestinmintempC', 
                                                  'totalsnow_cm':'Destintotalsnow_cm', 'sunhour':'Destinsunhour',
       'uvIndex':'DestinuvIndex', 'moon_illumination':'Destinmoon_illumination', 'moonrise':'Destinmoonrise', 
                                                  'moonset':'Destinmoonset', 'sunrise':'Destinsunrise',
       'sunset':'Destinsunset', 'dewpointC':'DestindewpointC', 'feelslikeC':'DestinfeelslikeC', 'heatindexC':'DestinheatindexC',
                                                  'windchillC':'DestinwindchillC',
       'windgustKmph':'DestinwindgustKmph', 'cloudcover':'Destincloudcover', 'humidity':'Destinhumidity', 
                                                  'precipMM':'DestinprecipMM', 'pressure':'Destinpressure',
       'tempC':'DestintempC', 'visibility':'Destinvisibility', 'winddirectiondegree':'Destinwinddirectiondegree', 
                                                  'windspeedKmph':'DestinwindspeedKmph' })
```

```python
#drop unnecessary features
origin_columns_drop = ['OriginuvIndex','Originmoon_illumination','Originmoonrise','Originmoonset','Originsunrise',
                    'Originsunset', 'OrigindewpointC','Originpressure','Originwinddirectiondegree']

destin_columns_drop = ['DestinuvIndex','Destinmoon_illumination','Destinmoonrise','Destinmoonset','Destinsunrise',
                    'Destinsunset', 'DestindewpointC','Destinpressure','Destinwinddirectiondegree' ]

origin_jan_2018.drop(columns = origin_columns_drop, inplace = True)

origin_jan_2019.drop(columns = origin_columns_drop, inplace = True)

destin_jan_2018.drop(columns = destin_columns_drop, inplace = True)

destin_jan_2019.drop(columns = destin_columns_drop, inplace = True)
```

```python
#merging origin and destination

origin_jan_weather = pd.concat([origin_jan_2018,origin_jan_2019])
origin_jan_weather = origin_jan_weather.rename(columns={ 'date_time':'fl_date', 'city':'origin_city_name'})
origin_jan_weather['fl_date'] = pd.to_datetime(origin_jan_weather['fl_date'])

destin_jan_weather = pd.concat([destin_jan_2018,destin_jan_2019])
destin_jan_weather = destin_jan_weather.rename(columns={ 'date_time':'fl_date', 'city':'dest_city_name'})
destin_jan_weather['fl_date'] = pd.to_datetime(destin_jan_weather['fl_date'])

data['fl_date'] = pd.to_datetime(data['fl_date'])
```

```python
# merging origin_jan_weather and data
df_merged = pd.merge(data, origin_jan_weather, left_on=['fl_date','origin_city_name'],right_on=['fl_date','origin_city_name'], how = 'inner')
```

```python
# merging destin_jan_weather and data
df_merged_final = pd.merge(df_merged, destin_jan_weather, left_on=['fl_date','dest_city_name'],right_on=['fl_date','dest_city_name'], how = 'inner')
```
